```python
class Point:
    def __init__(self, x,y):
        self.x = x
        self.y = y
```

## Helper functions


```python
import math
import operator

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size),0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)

def hsv_threshold(img,value_threshold):
    imagehsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    h = imagehsv[:,:,0]
    s = imagehsv[:,:,1]
    v = imagehsv[:,:,2]
    ret,v =  cv2.threshold(v, value_threshold, 255, cv2.THRESH_BINARY)
    return v

def detect_lines(img):
    # Detect lines in the resultant image
    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 80     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 5 #minimum number of pixels making up a line
    max_line_gap = 1    # maximum gap in pixels between connectable line segments
    line_image = np.copy(img)*0 # creating a blank to draw lines on

    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]),
                                min_line_length, max_line_gap)
    return lines

def post_process_edge_img(edge_img, dilate_threshold, erode_threshold):
    
    #Erode and dilate edges
    kernel_dilate = np.ones((dilate_threshold,dilate_threshold),np.uint8)
    kernel_erode = np.ones((erode_threshold,erode_threshold),np.uint8)
    
    edge_img = cv2.dilate(edge_img,kernelDilate)
    edge_img = cv2.erode(edge_img,kernelErode)
    
    return edge_img

def line_slope(line):
    for x1,y1,x2,y2 in line:
        slope = (y2-y1)/(x2-x1)    
        return slope


    
    
def largest_line(lines):
    line_count = len(lines)
    line_lengths = np.zeros(line_count, dtype=float) 
    line_id = 0
    for line in lines:
        for x1,y1,x2,y2 in line:
            line_lengths[line_id] = np.sqrt((x2-x1)*(x2-x1) + (y2-y1)*(y2-y1))
        line_id = line_id + 1
        
    max_index, max_value = max(enumerate(line_lengths), key=operator.itemgetter(1))
    return max_index,lines[max_index]

def connect_lines(lines, threshold):
    index, large_line = largest_line(lines)
    connected_line = large_line
    line_id = 0
    for line in lines:
        if(line_id==index):
            line_id = line_id+1
            continue
        for x1,y1,x2,y2 in line:
            pt1 = Point(x1,y1)
            pt2 = Point(x2,y2)
            d1 = point_line_distance(connected_line,pt1)
            d2 = point_line_distance(connected_line,pt2)
            if(d1<threshold and d2<threshold):
                connected_line = join_line(connected_line,line)
        line_id = line_id+1
    
    return connected_line
    

def join_line(connected_line,line):
    max_y_pt = []
    min_y_pt = []
    for x1,y1,x2,y2 in connected_line:
        if(y2>y1):
            max_y_pt = Point(x2,y2)
            min_y_pt = Point(x1,y1)
            connected_line = np.array([[x1,y1,x2,y2]])
        else:
            max_y_pt = Point(x1,y1)
            min_y_pt = Point(x2,y2)
            connected_line = np.array([[x2,y2,x1,y1]])

    for x1,y1,x2,y2 in line:
        if(y1>max_y_pt.y):
            connected_line = np.array([[connected_line[0][0],connected_line[0][1],x1,y1]])
            max_y_pt = Point(x1,y1)
        if(y2>max_y_pt.y):
            connected_line = np.array([[connected_line[0][0],connected_line[0][1],x2,y2]])
            max_y_pt = Point(x2,y2)
        if(y1<min_y_pt.y):
            connected_line = np.array([[x1,y1,connected_line[0][2],connected_line[0][3]]])
            min_y_pt = Point(x1,y1)
        if(y2<min_y_pt.y):
            connected_line = np.array([[x2,y2,connected_line[0][2],connected_line[0][3]]])
            min_y_pt = Point(x2,y2)
    return connected_line
    
def point_line_distance(line, point):
    for x1,y1,x2,y2 in line:
        p1 = np.array([x1, y1])
        p2 = np.array([x2, y2])
        p3 = np.array([point.x, point.y])
        d = np.linalg.norm(np.cross(p2-p1, p1-p3))/np.linalg.norm(p2-p1)
        return d
    
def extend_line(line):
    for x1,y1,x2,y2 in line:
        return
        
    
```

## Lane finding pipeline
 Below are the sequence of steps involved in the pipeline
 1. Convert image to HSV
 2. Apply Gaussian blur to *Value* channel of HSV 
 3. Apply threshold to the resulting image
 4. Apply Canny edge operator on the above image
 5. Apply mask to the above image to look for lanes only in specific area of the image
 6. Detect lines on the resultant output
 7. Connect lines in the image

```python
def detect_lanes(img,threshold, mask_vertices):
    imshape = image.shape
    # Convert image to HSV and take value component of image
    hsvimg = hsv_threshold(img, threshold)
    
    #Apply Gaussian Blur to the image
    blurimg = gaussian_blur(hsvimg,3)
    
    # Detect edges using Canny edge detector
    # Define our parameters for Canny and apply
    low_threshold = 50
    high_threshold = 150
    edges = cv2.Canny(blurimg, low_threshold, high_threshold)

    # Create and appy mask
    # Next we'll create a masked edges image using cv2.fillPoly()
    mask = np.zeros_like(edges)   
    ignore_mask_color = 255   
    # This time we are defining a four sided polygon to mask

    cv2.fillPoly(mask, mask_vertices, ignore_mask_color)
    masked_edges = cv2.bitwise_and(edges, mask)


    #Erode and dilate edges
    kernelDilate = np.ones((7,7),np.uint8)
    kernelErode = np.ones((5,5),np.uint8)

    masked_edges = cv2.dilate(masked_edges,kernelDilate)
    masked_edges = cv2.erode(masked_edges,kernelErode)

    # Detect lines in the resultant image
    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 80     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 5 #minimum number of pixels making up a line
    max_line_gap = 1    # maximum gap in pixels between connectable line segments
    line_image = np.copy(img)*0 # creating a blank to draw lines on

    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                                min_line_length, max_line_gap)

    positive_slope_lines = []
    negative_slope_lines = []
    
     # Iterate over the output "lines" and draw lines on a blank image
    line_id = 0
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = line_slope(line)
            if(slope>0):
                positive_slope_lines.append(line)
            else:
                negative_slope_lines.append(line)
    
    if(len(negative_slope_lines)):
        connected_line = connect_lines(negative_slope_lines,15)
        #final_line = extend_line(connected_line)
        for x1,y1,x2,y2 in connected_line:
            cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),10)
    
    if(len(positive_slope_lines)):
        connected_line = connect_lines(positive_slope_lines,15)
        #final_line = extend_line(connected_line)
        for x1,y1,x2,y2 in connected_line:
            cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),10)

    # Draw the lines on the edge image
    lines_edges = cv2.addWeighted(img, 0.8, line_image, 1, 0) 
    return lines_edges
```
